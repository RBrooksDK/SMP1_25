```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for Gaussian distribution
x = np.linspace(-3.5, 3.5, 500)
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Create the plot
fig, ax = plt.subplots(figsize=(8, 5), dpi=300)

# Plot Gaussian curve and shaded areas
ax.plot(x, y, color="#91c776", linewidth=2)
ax.fill_between(x, y, 0, where=(x < -2), color="#91c776", alpha=0.3)
ax.fill_between(x, y, 0, where=(x > 2), color="#91c776", alpha=0.3)

# Customise axes
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
#plt.arrow(3.5, 0, 0.3, 0, head_width=0.02, head_length=0.2, fc="black", ec="black")  # Arrow for x-axis
#plt.arrow(0, 0.4, 0, 0.03, head_width=0.1, head_length=0.02, fc="black", ec="black")  # Arrow for y-axis

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Set axis limits and labels
plt.xlim(-3.5, 3.5)
plt.ylim(0, 0.5)
plt.xticks([-3, -2, -1, 0, 1, 2, 3])
plt.yticks([])

# Remove background and save as transparent PNG
plt.gca().set_facecolor((0, 0, 0, 0))
plt.gcf().patch.set_alpha(0)
plt.savefig("gaussian.png", transparent=True, bbox_inches='tight', pad_inches=0)

plt.show()
```

```python
from sympy import symbols, diff, exp, log, solve

# Define variables
theta, n = symbols("theta n", positive=True, real=True)
x_vals = symbols("x1 x2 x3 x4 x5")  # Sample values

# Define the PDF for one observation
x = symbols("x", positive=True, real=True)
f_x = (1 / (2 * theta**3)) * x**2 * exp(-x / theta)

# Log-likelihood function for n observations
log_likelihood = sum(log(f_x.subs(x, x_val)) for x_val in x_vals)

# Derivative with respect to theta
dL_dtheta = diff(log_likelihood, theta)

# Solve for theta that maximizes the log-likelihood
theta_mle = solve(dL_dtheta, theta)
theta_mle
```
