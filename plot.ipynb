```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for Gaussian distribution
x = np.linspace(-3.5, 3.5, 500)
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Create the plot
fig, ax = plt.subplots(figsize=(8, 5), dpi=300)

# Plot Gaussian curve and shaded areas
ax.plot(x, y, color="#91c776", linewidth=2)
ax.fill_between(x, y, 0, where=(x < -2), color="#91c776", alpha=0.3)
ax.fill_between(x, y, 0, where=(x > 2), color="#91c776", alpha=0.3)

# Customise axes
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
#plt.arrow(3.5, 0, 0.3, 0, head_width=0.02, head_length=0.2, fc="black", ec="black")  # Arrow for x-axis
#plt.arrow(0, 0.4, 0, 0.03, head_width=0.1, head_length=0.02, fc="black", ec="black")  # Arrow for y-axis

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Set axis limits and labels
plt.xlim(-3.5, 3.5)
plt.ylim(0, 0.5)
plt.xticks([-3, -2, -1, 0, 1, 2, 3])
plt.yticks([])

# Remove background and save as transparent PNG
plt.gca().set_facecolor((0, 0, 0, 0))
plt.gcf().patch.set_alpha(0)
plt.savefig("gaussian.png", transparent=True, bbox_inches='tight', pad_inches=0)

plt.show()
```
